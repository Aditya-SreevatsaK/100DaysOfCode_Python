```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Analyse the 2018 Central Park Squirrel Census Data set and create a CSV file containing the number of each type of colored squirrel.

```python
squirrel_data = pd.read_csv("2018_Central_Park_Squirrel_Census_-_Squirrel_Data.csv")
squirrel_data
```

```python
primary_fur_color = "Primary Fur Color"
squirrel_black_fur = len(squirrel_data[squirrel_data[primary_fur_color] == "Black"])
squirrel_gray_fur = len(squirrel_data[squirrel_data[primary_fur_color] == "Gray"])
squirrel_cinnamon_fur = len(squirrel_data[squirrel_data[primary_fur_color] == "Cinnamon"])

squirrel_color_data = {
    "Primary Fur Color" : ["Black", "Gray", "Cinnamon"],
    "Count of Squirrels" : [squirrel_black_fur, squirrel_gray_fur, squirrel_cinnamon_fur]
}

squirrel_color_df = pd.DataFrame(squirrel_color_data).to_csv("Squirrel Count By Colour.csv")
```

```python
plt.figure(figsize = (5, 5))
plot = sns.countplot(squirrel_data, x = primary_fur_color)
plt.ylabel("Count of Squirrels")
plt.title("Count of Squirrels of different colors.")
plt.show()
```
